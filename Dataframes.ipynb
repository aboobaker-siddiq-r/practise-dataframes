```python
import matplotlib.pyplot as plt 
import seaborn as sns
plt.rcParams['figure.figsize']=8,4
# %matplotlib inline
```

```python
stats.head()
```

```python
# sns.set(rc = {'figure.figsize':(20,20)})
vis1=sns.displot(stats["InternetUsers"],bins=10)
# plt.ylim(5, 45)
```

```python
vis2=sns.boxplot(data=stats,x='IncomeGroup',y='BirthRate')
```

```python
import seaborn as sns
sns.set_theme(style="darkgrid")
sns.lineplot(x='IncomeGroup', y='BirthRate',
             data=stats)
```

```python
sns.violinplot(x='IncomeGroup', y='BirthRate',
             data=stats)
```

```python
sns.lmplot(x='InternetUsers',y='BirthRate',data=stats,fit_reg=False,hue='IncomeGroup',size=10,aspect=1)
```

```python
#increasing the marker size
sns.lmplot(x='InternetUsers',y='BirthRate',data=stats,fit_reg=False,hue='IncomeGroup',size=8,scatter_kws={"s":200})
```
